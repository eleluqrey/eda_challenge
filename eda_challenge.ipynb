```python
diamonds_transactional = pd.read_csv('csv_tables/diamonds_transactional.csv')
diamonds_transactional
```

```python
# ANALIZA EL DATA FRAME COMPLETO

# 1ยบ DATA FRAME COMPLETO
# Hacemos merge para unir todos los data frames de par en par uniendo por la columna que tengan en comรบn.

diamonds_df = diamonds_color.merge(diamonds_properties,on = 'color_id')
diamonds_df
```

```python
diamonds_df = diamonds_df.merge(diamonds_cut,on = 'cut_id')
diamonds_df
```

```python
diamonds_df = diamonds_df.merge(diamonds_clarity,on = 'clarity_id')
diamonds_df
```

```python
diamonds_df = diamonds_df.merge(diamonds_transactional,on = 'index_id')
diamonds_df
```

```python
diamonds_df = diamonds_df.merge(diamonds_city,on = 'city_id')
diamonds_df
```

```python
diamonds_df = diamonds_df.merge(diamonds_dimensions,on = 'index_id')
diamonds_df
```

```python
# 2ยบ Una vez tenemos el data frame completo con toda la informaciรณn vamos a analizarlo:
```

```python
# Exploring numeric features
@interact(Feature=list(diamonds_df.select_dtypes(include=['float64','int64']).columns))

def histogram(Feature):
    fig = px.histogram(diamonds_df[Feature], marginal = 'box')
    fig.show()
```

```python
fig = px.histogram(diamonds_df['depth'], marginal = 'box')
fig.show()
```

```python
@interact(Feature=list(diamonds_df.select_dtypes(include=['object']).columns[1:]))

def histogram(Feature):
    fig = px.bar(diamonds_df[Feature])
    fig.show()
```

```python
corr_df = diamonds_df.select_dtypes(include = ['int64','float64']).corr()
px.imshow(corr_df, text_auto = True)
```
